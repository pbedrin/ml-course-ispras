**Задача 1. (4 балла)**

Найдите $f'(t)$ и $f''(t)$ функции $f: \mathbb{R} \to \mathbb{R}$:

$f(t) = \det(A - tI_d), A \in \mathbb{S}^d$, при $\det(A - tI_d) \neq 0$.

*Решение.*

Обозначим $B(t) = A - tI_d, \quad f(t) = \det(B(t)) $.

$df(t) = d(\det(B(t))) = \det(B(t)) \cdot \mathrm{Tr}(B^{-1}(t) \cdot dB(t))$

* $dB(t) = d(A - tI_d) = dA - d(tI_d) = -I_d dt$
* $\mathrm{Tr}(B^{-1}(t) \cdot dB(t)) = -\mathrm{Tr}(B^{-1}(t) \cdot I_d dt) = -\mathrm{Tr}(B^{-1}(t))dt $

Получается: $ df(t) = - \det(B(t)) \cdot \mathrm{Tr}(B^{-1}(t)) \cdot dt $

Первая производная: $f'(t) = \frac{df(t)}{dt} = - \det(B(t)) \cdot \mathrm{Tr}(B^{-1}(t)) = - \det(A - tI_d) \cdot \mathrm{Tr}((A - tI_d)^{-1}).$

$ d^2 f(t) = d(df(t)) = -d(\det(B(t)) \cdot \mathrm{Tr}(B^{-1}(t)) \cdot dt) = -d(\det(B(t)))\cdot \mathrm{Tr}(B^{-1}(t)) - \det(B(t)) \cdot d(\mathrm{Tr}(B^{-1}(t))) $

Нужно вычислить:
* $d(\mathrm{Tr}(B^{-1}(t))) = \mathrm{Tr}(dB^{-1}(t)) = \mathrm{Tr}(-B^{-1}(t) \cdot dB(t) \cdot B^{-1}(t)) = \mathrm{Tr}(B^{-2}(t)) dt$

Подставим $d(\mathrm{Tr}(B^{-1}(t)))$ и $d(\det(B(t)))$ в $d^2 f$:

$ d^2 f(t) = \det(B(t)) \cdot \mathrm{Tr}(B^{-1}(t)) \cdot dt \cdot \mathrm{Tr}(B^{-1}(t)) - \det(B(t)) \cdot \mathrm{Tr}(B^{-2}(t)) dt $

Тогда $f''(t) = \frac{d^2 f(t)}{dt^2} = \det(B(t)) \cdot \mathrm{Tr^2}(B^{-1}(t)) - \det(B(t)) \cdot \mathrm{Tr}(B^{-2}(t)) = \det(A - tI_d) [ \mathrm{Tr}^{2}((A - tI_d)^{-1}) - \mathrm{Tr}((A - tI_d)^{-2})] $

Ответ:
* $ f'(t) = - \det(A - tI_d) \cdot \mathrm{Tr}((A - tI_d)^{-1}) $
* $ f''(t) = \det(A - tI_d) [ \mathrm{Tr}^{2}((A - tI_d)^{-1}) - \mathrm{Tr}((A - tI_d)^{-2})] $

**Задача 2. (8 баллов)**

Найдите $\nabla f(x)$ и $\nabla^2 f(x)$ функции $f: \mathbb{R}^d \to \mathbb{R}$:

**2.1. (4 балла):** $f(x) = \frac{1}{2} \lVert A - xx^\top \rVert_F^2, \quad A \in \mathbb{S}^d.$

*Решение.*

Обозначим $B(x) = A - xx^\top$. 

$df(x) = d(\frac{1}{2}\|B(x)\|_F^2) = \frac{1}{2}d\langle B(x), B(x)\rangle = \frac{1}{2} \cdot 2\langle B(x), dB(x)\rangle = \langle B(x), dB(x)\rangle$.

$dB(x) = d(A - xx^\top) = -d(xx^\top) = -(dx)x^\top - x(dx^\top) = -(dx)x^\top - x(dx)^\top$, подставим:

$df(x) = \langle B(x), -(dx)x^\top - x(dx)^\top\rangle = -\langle B(x), (dx)x^\top\rangle - \langle B(x), x(dx)^\top\rangle = -\langle B(x)x, dx\rangle - \langle B(x)dx, x\rangle = $

$ = -\langle B(x)x, dx\rangle - \langle B^\top(x)x, dx\rangle = -2\langle B(x)x, dx\rangle = \langle -2B(x)x, dx\rangle = \langle -2(A - xx^\top)x, dx\rangle$

Так как $df(x) = \langle \nabla f(x), dx\rangle$, то $\nabla f(x) = -2(A - xx^\top)x$.

Теперь найдём гессиан.

$d(\nabla f(x)) = d(-2(A - xx^\top)x) = -2[d(A - xx^\top)x + (A - xx^\top)dx] = -2[( -(dx)x^\top - x(dx)^\top ) x + (A - xx^\top)dx] = $

$ = -2[-(dx)x^\top x - x(dx)^\top x + (A - xx^\top)dx] = -2[-x^\top xI_d - xx^\top + A - xx^\top]dx = $

$= [2x^\top xI_d + 4xx^\top - 2A]dx = J_{\nabla f}dx$.

$\nabla^2 f = (J_{\nabla f})^\top. \quad $Из симметричности $J_{\nabla f}: \nabla^2 f = 2x^\top xI_d + 4xx^\top - 2A$.

Ответ:

* $ \nabla f(x) = -2(A - xx^\top)x $
* $ \nabla^2 f(x) = 2x^\top xI_d + 4xx^\top - 2A $

**2.2. (4 балла):** $ f(x) = \langle x, x \rangle ^{\langle x, x \rangle}, $ при $ x \neq 0 $

*Решение.*

Введём замену: $ t = \langle x, x \rangle, \quad \langle x, x \rangle ^{\langle x, x \rangle} = t^t = (e^{\ln t})^t = e^{t \ln t}, \quad dt = d(\langle x, x \rangle) = 2 \langle x, dx \rangle$

Обозначим $ g(t) = t \ln t$, тогда $f(g(t)) = e^{g(t)} $

$ d(f(g(t))) = d(e^{g(t)}) = e^{g(t)} \cdot dg(t) = e^{t \ln t} \cdot d(t \ln t)$,

$ \frac{d(t \ln t)}{dt} = \frac{t \cdot d(\ln t) + (\ln t)dt}{dt} = \frac{t \frac{1}{t} dt + \ln t \cdot dt}{dt} = 1 + \ln t, \quad d(t \ln t) = (1 + \ln t)dt $, подставим в выражение выше:

$ d(f(g(t))) = e^{t \ln t} \cdot (1 + \ln t) \cdot dt = t^t \cdot (1 + \ln t) \cdot dt $

Обратная замена.

$ df(x) = 2 \langle x, x \rangle ^{\langle x, x \rangle} (1 + \ln \langle x, x \rangle) \langle x, dx \rangle = \langle 2 \langle x, x \rangle ^{\langle x, x \rangle} (1 + \ln \langle x, x \rangle) x, dx \rangle$

Так как $df(x) = \langle \nabla f(x), dx\rangle$, то $ \nabla f(x) = 2 \langle x, x \rangle ^{\langle x, x \rangle} (1 + \ln \langle x, x \rangle) x$.

Теперь найдём гессиан.

$\nabla f(x) = g(x) \cdot x$, где $g(x) = 2 \langle x, x \rangle ^{\langle x, x \rangle} (1 + \ln \langle x, x \rangle) $.

Вернёмся для удобства вновь к замене с $t$:

$ g(t) = 2 t^{t} (1 + \ln t) = 2 e^{t \ln t} (1 + \ln t) $

$ dg(t) = 2 d[e^{t \ln t} (1 + \ln t)] = 2 [d(1 + \ln t) \cdot e^{t \ln t} + (1 + \ln t) \cdot d(e^{t \ln t})] $

Найдём и подставим:

* $ d(1 + \ln t) = \frac{dt}{t} $

* $ d(e^{t \ln t}) = e^{t \ln t} \cdot d(t \ln t) = e^{t \ln t} \cdot (1 + \ln t) dt$

$ dg(t) = 2 [\frac{dt}{t} \cdot e^{t \ln t} + (1 + \ln t) \cdot e^{t \ln t} \cdot (1 + \ln t) dt] = 2 e^{t \ln t} [\frac{1}{t} + (1 + \ln t)^2] dt $

$ dg(x) = 4 e^{\langle x, x \rangle \ln \langle x, x \rangle} [\frac{1}{\langle x, x \rangle} + (1 + \ln \langle x, x \rangle)^2 ] \langle x, dx \rangle $

$ d(\nabla f(x)) = dg(x) \cdot x + g(x) \cdot dx $

$ d(\nabla f(x)) = 4 e^{\langle x, x \rangle \ln \langle x, x \rangle} [\frac{1}{\langle x, x \rangle} + (1 + \ln \langle x, x \rangle)^2 ] \langle x, dx \rangle x + $
$ 2 \langle x, x \rangle ^{\langle x, x \rangle} (1 + \ln \langle x, x \rangle) dx = $

$ = [ 4 \langle x, x \rangle^{\langle x, x \rangle} ( \frac{1}{\langle x, x \rangle} + (1 + \ln \langle x, x \rangle)^2 ) xx^\top + 2 \langle x, x \rangle^{\langle x, x \rangle} (1 + \ln \langle x, x \rangle) I_d ] dx = J_{\nabla f}dx $

$\nabla^2 f = (J_{\nabla f})^\top. \quad $Из симметричности $J_{\nabla f}: \nabla^2 f = 4 \langle x, x \rangle^{\langle x, x \rangle} ( \frac{1}{\langle x, x \rangle} + (1 + \ln \langle x, x \rangle)^2 ) xx^\top + 2 \langle x, x \rangle^{\langle x, x \rangle} (1 + \ln \langle x, x \rangle) I_d $

Ответ:
* $ \nabla f(x) = 2 \langle x, x \rangle ^{\langle x, x \rangle} (1 + \ln \langle x, x \rangle) x $
* $ \nabla^2 f(x) = 4 \langle x, x \rangle^{\langle x, x \rangle} ( \frac{1}{\langle x, x \rangle} + (1 + \ln \langle x, x \rangle)^2 ) xx^\top + 2 \langle x, x \rangle^{\langle x, x \rangle} (1 + \ln \langle x, x \rangle) I_d $

**Задача 3. (8 баллов)**

Найдите $\nabla f(X)$ и $d^2f(X)$ функции $f: \mathbb{S}_{++}^d \to \mathbb{R}$:

**3.1. (4 балла):**

$ f(X) = \langle (A + X)^2, (BX^{-1} + C) \rangle, \quad A, B, C \in \mathbb{R}^{d \times d}$.

*Решение.*

$ f(X) = \langle (A+X)^2, (BX^{-1} + C) \rangle = \langle (A+X)^2, BX^{-1} \rangle + \langle (A+X)^2, C \rangle$

$ df(X) = d\langle (A+X)^2, BX^{-1} \rangle + d\langle (A+X)^2, C \rangle $

* $ d\langle (A+X)^2, BX^{-1} \rangle = \langle d(A+X)^2, BX^{-1} \rangle + \langle (A+X)^2, d(BX^{-1}) \rangle$

* $ d\langle (A+X)^2, C \rangle = \langle d(A+X)^2, C \rangle $

Вычислим дифференциалы, используемые выше:

* $ d(A+X)^2 = (d(A+X))(A+X) + (A+X)d(A+X) = dX(A+X) + (A+X)dX $

* $ d(BX^{-1}) = -BX^{-1}(dX)X^{-1} $

Подставляем:

* $ d\langle (A+X)^2, BX^{-1} \rangle = \langle d(A+X)^2, BX^{-1} \rangle + \langle (A+X)^2, d(BX^{-1}) \rangle = \langle (dX)(A+X), BX^{-1} \rangle + \langle (A+X)dX, BX^{-1} \rangle + \langle (A+X)^2, -BX^{-1}(dX)X^{-1} \rangle $

* $ d\langle (A+X)^2, C \rangle = \langle d(A+X)^2, C \rangle = \langle (dX)(A+X), C \rangle + \langle (A+X)dX, C \rangle $

Преобразуем скалярные произведения, чтобы отделить $dX$:

* $ \langle (dX)(A+X), BX^{-1} \rangle = \langle BX^{-1}(A+X)^T, dX \rangle $

* $ \langle (A+X)dX, BX^{-1} \rangle = \langle (A+X)^TBX^{-1}, dX \rangle $

* $ \langle (A+X)^2, -BX^{-1}(dX)X^{-1} \rangle = -\langle X^{-1} B^\top (A+X)^2 X^{-1}, dX \rangle$

* $ \langle (dX)(A+X), C \rangle = \langle C(A+X)^T, dX \rangle$

* $ \langle (A+X)dX, C \rangle = \langle (A+X)^TC, dX \rangle$

$ df(X) = \langle BX^{-1}(A+X)^T, dX \rangle + \langle (A+X)^TBX^{-1}, dX \rangle - \langle X^{-1}B^T(A+X)^2X^{-1}, dX \rangle + \langle C(A+X)^T, dX \rangle + \langle (A+X)^TC, dX \rangle$

$ df(X) = \langle BX^{-1}(A+X)^T + (A+X)^TBX^{-1} - X^{-1}B^T(A+X)^2X^{-1} + C(A+X)^T + (A+X)^TC, dX \rangle$

$df(X) = \langle \nabla f(X), dX\rangle$. Тогда:

$ \nabla f(X) = BX^{-1}(A+X)^T + (A+X)^TBX^{-1} - X^{-1}B^T(A+X)^2X^{-1} + C(A+X)^T + (A+X)^TC $

Перейдем к поиску $d^2 f(X)$. Приращение $dX$ фиксируем как $dX_1$. 

$df(X) = \langle \nabla f(X), dX \rangle, \quad d^2 f(X) = \langle d(\nabla f(X))[dX_2], dX_1 \rangle $

* $d(BX^{-1}(A+X)^T)[dX_2] = B(-X^{-1} dX_2 X^{-1})(A+X)^T + BX^{-1} dX_2^T$
* $d((A+X)^T B X^{-1})[dX_2] = dX_2^T B X^{-1} + (A+X)^T B (-X^{-1} dX_2 X^{-1})$
* $d(-X^{-1} B^T (A+X)^2 X^{-1})[dX_2] = X^{-1} dX_2 X^{-1} B^T (A+X)^2 X^{-1} - X^{-1} B^T (dX_2 (A+X) + (A+X) dX_2) X^{-1} + X^{-1} B^T (A+X)^2 X^{-1} dX_2 X^{-1}$
* $d(C (A+X)^T)[dX_2] = C \cdot dX_2^T$
* $d((A+X)^T C)[dX_2] = dX_2^T C$

Тогда, подставив...

$ d^2 f(X) = $
$B(-X^{-1} dX_2 X^{-1})(A+X)^T + BX^{-1} dX_2^T $
$+ dX_2^T B X^{-1} + (A+X)^T B (-X^{-1} dX_2 X^{-1}) $
$+ X^{-1} dX_2 X^{-1} B^T (A+X)^2 X^{-1}  $
$- X^{-1} B^T \left(dX_2 (A+X) + (A+X) dX_2\right) X^{-1} $
$+ X^{-1} B^T (A+X)^2 X^{-1} dX_2 X^{-1}$
$+ C dX_2^T + dX_2^T C$

Ответ:
* $\nabla f(X) = BX^{-1}(A+X)^T + (A+X)^TBX^{-1} - X^{-1}B^T(A+X)^2X^{-1} + C(A+X)^T + (A+X)^TC $
* $d^2 f(X) = \langle d(\nabla f(X))[dX_2], dX_1 \rangle$ (полностью - в решении)

**3.2. (4 балла):**

$f(X) = (\det(X))^{1/d}, \ d \in \mathbb{R}.$

*Решение:*

$\ln f(X) = \ln ((\det(X))^{1/d}) = \frac{1}{d} \ln(\det(X)) $

$ f(X) = \exp({\frac{1}{d} \ln \det(X)}) $

$ df(X) = f(X) \cdot \frac{1}{d} \cdot d( \ln(\det(X))) = f(X) \cdot \frac{1}{d} \cdot \frac{d(\det(X))}{\det(X)} = f(X) \cdot \frac{1}{d} \cdot \frac{\det(X) \cdot \mathrm{Tr}(X^{-1} dX)}{\det(X)} = f(X) \cdot \frac{1}{d} \cdot \mathrm{Tr}(X^{-1} dX) = $
$ f(X) \cdot \frac{1}{d} \cdot \langle X^{-1}, dx \rangle = \langle \frac{1}{d} (\det(X))^{1/d} X^{-1}, dx \rangle $

$ \nabla f(X) = \langle df(X), dX \rangle$. Тогда:

$ \nabla f(X) = \frac{1}{d} (\det(X))^{1/d} X^{-1} $

Перейдем к поиску $d^2 f(X)$. Приращение $dX$ фиксируем как $dX_1$. 

$d^2 f(X) = d[f(X) \cdot \frac{1}{d} \cdot \mathrm{Tr}(X^{-1} dX_1)] = \frac{1}{d} \cdot d[f(X) \cdot \mathrm{Tr}(X^{-1} dX_1)] = $
$ \frac{1}{d} \cdot [d(f(X)) \cdot \mathrm{Tr}(X^{-1} dX_1) + f(X) \cdot d(\mathrm{Tr}(X^{-1} dX_1)) ]$

* $ df(X) = f(X) \cdot \frac{1}{d} \cdot \mathrm{Tr}(X^{-1} dX_2) $
* $ d(\mathrm{Tr}(X^{-1} dX_1)) = \mathrm{Tr}(d(X^{-1} dX_1)) = \mathrm{Tr}(-X^{-1}(dX_2)X^{-1} \cdot dX_1) $

Подставим:
$ d^2 f(X) = \frac{1}{d^2} f(X) \cdot \mathrm{Tr}(X^{-1} dX_2) \cdot \mathrm{Tr}(X^{-1} dX_1) - \frac{1}{d} f(X) \cdot \mathrm{Tr}(X^{-1} dX_2 X^{-1} dX_1) $

Ответ:
* $\nabla f(X) = \frac{1}{d} (\det(X))^{1/d} X^{-1} $
* $ d^2 f(X) = \frac{1}{d^2} f(X) \cdot \mathrm{Tr}(X^{-1} dX_2) \cdot \mathrm{Tr}(X^{-1} dX_1) - \frac{1}{d} f(X) \cdot \mathrm{Tr}(X^{-1} dX_2 X^{-1} dX_1) $
